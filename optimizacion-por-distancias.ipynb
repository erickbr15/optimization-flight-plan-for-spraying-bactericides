```r
#Función para el calculo de delta por distancias
compute.delta.of.jump <- function(working.idx, jump.length){
  
  distancias.working.idx <- m_adyacencia[working.idx,]
  
  min.distance.of.row <- min(distancias.working.idx[distancias.working.idx > 0])
    
  delta <- jump.length / min.distance.of.row
  
  return (delta)
}
```

```r
#Funciones para la evaluación de la distancia propuesta en el cromosoma binario

find.the.next.jump <- function(working.cromosoma, working.idx){
  
  cromosoma.distancias <- working.cromosoma * m_adyacencia[working.idx,]
  
  if(!any(cromosoma.distancias > 0)){
    return (list(can.jump <- FALSE, step.0 <- working.idx, step.1 <- working.idx, distance <- 0))
  }
  
  distancia_min <- min(cromosoma.distancias[cromosoma.distancias > 0])
  new.working.idx <- match(distancia_min, cromosoma.distancias)
  
  return (list(can.jump <- TRUE, step.0 <- working.idx, step.1 <- new.working.idx, distance <- distancia_min))
}

compute.delta.distance <- function(working.cromosoma, verbose){
  
  CAN.JUMP.IDX <- 1
  STEP.0.IDX <- 2
  STEP.1.IDX <- 3
  DISTANCE.IDX <- 4
  
  working.idx <- 0
    
  for(i in 1:length(working.cromosoma)){            
    if(working.cromosoma[i] == 1){
      working.idx <- i      
      break
    }
  }
  
  if(working.idx == 0){
    return (0)
  }
  
  next.jump.result <- find.the.next.jump(working.cromosoma, working.idx)
  can.jump = next.jump.result[[CAN.JUMP.IDX]]
  
  if(can.jump == FALSE){
    if(verbose == TRUE){
      print(paste0(next.jump.result[[STEP.0.IDX]], " ,distance=NULL"))
    }
    return (0)
  }    
    
  while(can.jump == TRUE){        
    
    if(verbose == TRUE){
      print(paste0(next.jump.result[[STEP.0.IDX]], " - ", next.jump.result[[STEP.1.IDX]], " ,distance= ", next.jump.result[[DISTANCE.IDX]]))
    }
    
    delta <- compute.delta.of.jump(next.jump.result[[STEP.0.IDX]], next.jump.result[[DISTANCE.IDX]])
    if(delta > max.distance.delta.allowed){
        return(delta)
    }    
    
    working.cromosoma[next.jump.result[[STEP.0.IDX]]] <- 0
    
    next.jump.result <- find.the.next.jump(working.cromosoma, next.jump.result[[STEP.1.IDX]])
    can.jump = next.jump.result[[CAN.JUMP.IDX]]        
  }
  
  if(verbose == TRUE){
    print(paste0(next.jump.result[[STEP.0.IDX]], " - ", next.jump.result[[STEP.1.IDX]], " ,distance= ", next.jump.result[[DISTANCE.IDX]]))
  }
  
  return (0)
}
```

```r
#Implementación de la función fitness por area
compute.delta.area <- function(c){
    
  resultado <- 0  
  area.total <- c %*% parcelas.disponibles.df$area.cultivada  
  delta <- area.total / area.diaria.objetivo
  
  return(delta)  
}
```

```r
#Implementación de la función fitness por area y por distancia
fitness.area.and.distance <- function(c){
    
    delta.area <- compute.delta.area(c)
    
    if(delta.area < 0.995 || delta.area > 1.005){
        return(10000)
    }
        
    c.for.distance <- c
    
    if(length(c) < numero.total.parcelas){
        parcelas.data <- parcelas.disponibles.df %>% filter(id.parcela %in% parcelas.disponibles.df$id.parcela[c == 1])
        
        c.for.distance <- c(rep(0,numero.total.parcelas))
        
        for(i in 1:numero.total.parcelas){            
            id.parcela.target = i - 1
            if(any(parcelas.data$id.parcela == id.parcela.target)){
                c.for.distance[i] <- 1                
            }
        }        
    }
    
    delta.distance <- compute.delta.distance(c.for.distance, FALSE)
    
    if(delta.distance > max.distance.delta.allowed){
        return(10000)
    }
    
    resultado.fitness <- 0
    if (delta.area > 1.005) {
        resultado.fitness <- area.diaria.objetivo * exp(delta.area^3 - 1) 
    } 
    if (delta.area < .995) {
        resultado.fitness <- area.diaria.objetivo * exp(1 - delta.area^2) 
    }
    
    return(resultado.fitness)
}
```

```r
#Plan Día 3
best <- optimization.dia.3$population[optimization.dia.3$evaluations == min(optimization.dia.3$best),][1,]

area.cubierta.dia.3 <- best %*% parcelas.disponibles.df$area.cultivada
parcelas.dia.3 <- parcelas.disponibles.df %>% filter(id.parcela %in% parcelas.disponibles.df$id.parcela[best == 1])

print(paste0("El area cubierta el Día 3: ", area.cubierta.dia.3))
parcelas.dia.3
```
